# 2. Преобразование признаков

#### Признаки умножают на обратимую матрицу. Изменится ли качество линейной регрессии? (Её можно обучить заново.)

Рассмотрим работу линейной регрессии. У нас есть матрица признаков $X_{train}$ (при этом первый стобец состоит из единиц) и столбец целевых признаков $ y_{train} $ . 

Изходя из признаков и целевых признаков можем получить оптимальный вектор весов для минимизации функции потерь (MSE): $w = (X^{T}X)^{-1}X^{T}y$, MSE = $ \frac{1}{n} * ||Xw - y||_{d_2} $

Посмотрим, что будет, если мы умножим матрицу признаков на обратимую матрицу: 

$ X_{new} = X_{train}A $

$w_{new} = (X_{new}^{(T)} X_{new})^{-1}  X_{new}^{(T)} * y = $

$((X_{train}A)^{(T)} *   X_{train}A)^{-1} * (X_{train}A)^{(T)} * y = $

$(A^{(T)} * X_{train}^{(T)} X_{train} * A)^{-1}  * A^{(T)}X_{train}^{(T)} * y = $

$ A^{-1} * (X_{train}^{(T)} X_{train})^{-1} * (A^{(T)})^{-1}A^{(T)} * X_{train}^{(T)} * y = $

$ A^{-1} * (X_{train}^{(T)} X_{train})^{-1}X_{train}^{(T)} * y = A^{-1} * w$

$ predict_{new} = X_{new} * w_{new} = X_{train}A * A^{-1} * w = X_{train}w = predict_{old}$

Следовательно, от линейного преобразования матрицы признаков, качество модели не изменяется.

# 3 Алгоритм преобразования данных для решения задачи. 

1) Сначала добавим столбец единиц к матрице признаков 

2) Затем создадим случайную матрицу и умножим на нее матрицу X

```python
def get_matrix(X):
    bul = True
    n = X.shape[1]
    A = np.random.random((n,n))
    while(bul!=False): 
        try: 
            np.linalg.inv(A)
            bul = False
        except: 
            A = np.random.random((n,n))
            bul = True   
    return A

def upgrade(X, A): 
    matrix = X @ A
    return matrix
```

Качество линейной регрессии не поменяется, так как ее ответы будут совпадать с моделью, которая обрабатывает первоначальные признаки (см. вывод п.2)

# 4. Проверка качества модели. 

```python
RND = 123
```

```python
def transform_df(data):
    X = data.drop(['Страховые выплаты'], axis=1)
    y = data['Страховые выплаты']
    return X, y
```

```python
df_train, df_test = train_test_split(data, test_size=0.25, random_state=RND)
X_train, y_train = transform_df(df_train)
X_test, y_test = transform_df(df_test)

A = get_matrix(X_train)

X_train_new = upgrade(X_train, A)
X_test_new = upgrade(X_test, A)
```

```python
A
```

```python
model_1 = LinearRegression()
model_2 = LinearRegression()

model_1.fit(X_train, y_train)
model_2.fit(X_train_new, y_train)

y_pred_1 = model_1.predict(X_test)
y_pred_2 = model_2.predict(X_test_new)

score_1 = r2_score(y_pred_1, y_test)
print('R2_score для обычной модели:', score_1.round(3))

score_2 = r2_score(y_pred_2, y_test)
print('R2_score для модели с модифицированными признаками:', score_2.round(3))

print('Сравнение двух моделей между собой (R2_score):', r2_score(y_pred_1, y_pred_2))
```
